```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import os
from time import sleep
import time

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
```

```python
#this function get you the list of lyrics links
def get_lyric(ext_url):
    #artist = input ('enter url extension after lyrics.com/artist/ ')
    ext_url_request = requests.get(f'http://www.lyrics.com/artist/{ext_url}').text
    pattern = '<a href="/lyric/[^>]*>'
    songs_list = re.findall(pattern, ext_url_request)[1:101]
    
    links = []
    for r in songs_list:
        links.append('https://www.lyrics.com'+ r[9:])
    df = pd.DataFrame({'links': links})
    return df
```

```python
get_lyric('Lokua-Kanza/173548')
```

```python
def get_lyrics(ext_url, artist_name):
    base_url = 'http://www.lyrics.com/artist/'
    url = base_url + ext_url
    r = requests.get(url)
    soup = BeautifulSoup(r.text , 'html.parser')
    s = []
    l = []
    songs = soup.find_all('td', {'class': 'tal qx'})
    for i in songs:
        s.append(i.text)
    for i in range(len(songs)):
        if songs[i].a is None:
            pass
        else:
            lyr = base_url + songs[i].a.attrs['href']
            url = lyr
            r = requests.get(url)
            soup = BeautifulSoup(r.text , 'html.parser')
            l.append(soup.find('pre', {'id' : 'lyric-body-text'}).text)
    
    df = pd.DataFrame({'artist': artist_name, 'songs': s, 'lyrics': l})
    #df['lyrics'] = df['lyrics'].str.replace('\W', ' ').lower()
    return df
```

```python
get_lyrics('Lokua-Kanza/173548', 'Lokua-Kanza')
```

```python
get_lyrics('artist/Enya/4197', 'Enya')
```

```python
def get_lyrics2():
    artist = input ('who you wan listen? ')
    
    if len(artist.split()) < 2:
        url = 'http://www.lyrics.com/artist/' + artist.split(' ')[0].lower()
    else:
        url = 'http://www.lyrics.com/artist/' + artist.split(' ')[0].lower()
    
    for i in range (1, len(artist.split())):
        url = url + '%20' + artist.split(' ')[i].lower()
        
    r = requests.get(url)
    soup = BeautifulSoup(r.text , 'html.parser')
    alb =[]
    s = []
    t = []
    l = []
    albums = soup.find_all('h3', {'class': 'artist-album-label'})
    for i in albums:
        alb.append(i.text)
    songs = soup.find_all('td', {'class': 'tal qx'})
    for i in songs[::2]:
        s.append(i.text)
    for i in songs[1::2]:
        t.append(i.text)
    base_url = 'http://www.lyrics.com/'
    for i in range(len(songs)):
        if songs[i].a is None:
            pass
        else:
            lyr = base_url + songs[i].a.attrs['href']
            url = lyr
            r = requests.get(url)
            soup = BeautifulSoup(r.text , 'html.parser')
            l.append(soup.find('pre', {'id' : 'lyric-body-text'}).text)
    
    df = pd.DataFrame({'songs': s, 'time': t, 'lyrics': l})
    #df['lyrics'] = df['lyrics'].str.replace('\W', ' ').lower()
    return df
```
